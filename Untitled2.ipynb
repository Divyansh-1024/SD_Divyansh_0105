```python
l=[1,4,7,9,"clg","and",45,45.26,"for"]
```

```python
l1_num=[]
l2_str=[]
for i in l:
  if type(i)==int or type(i)==float:
    l1_num.append(i)
  else:
    l2_str.append(i)
```
